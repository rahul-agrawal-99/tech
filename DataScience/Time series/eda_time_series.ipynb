## Time resampling

```python
df["Open"].resample(rule="A").min()      # showing the minimum value of open price for each year  , rule = "A" means annual
```

<img src="https://miro.medium.com/max/1174/1*JKRKrjcDUMUIOTE6VOTwGQ.png">

```python
df["Close"].resample(rule="M").max()       # showing the maximum value of close price for each moonth  , rule = "M" means monthly
```

```python
df["Close"].resample(rule="M").max().plot().set_title("Monthly Max Close Price")
```

```python
# showing the mean value (at the end of quarter ) of high price for each quarter  , rule = "Q" means quarterly

#  Using bar chart to show the mean value of high price for each quarter
 
df["High"].resample(rule="Q").mean().plot(kind="bar").set_title("Quaterly Mean High Price")    
```

```python
#  getting a max of High price from entire dataset

df[df["High"] ==df["High"].max()]        
```

```python
#  using rolling function 

df.rolling(window=3).mean()     # get 3 values in each row and calculate the mean of that , so that first ( window_size - 1)  are null
```

```python
df[["Open","Close"]].plot().set_title("Open and Close Price combnined")
```
