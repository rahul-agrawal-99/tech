# Linux Directory Structure



## / => top level directory like c drive 
## /home => home dir for other users   e.g /home/user_name  will be home dir for user
## /root => home dir for root user
## /boot => os required files are placed in this dir , dont modify
## /etc =>(editable text configuartion ) it contains all configuration files , Files used by application subsystems
## /var => application software packages are installed by default here.  log files, locks, spool files, and temporary files stored here
## /bin => user level commands stored here by default , binary executables essential for entire os like ls , vi , ss , every user has seperate bin and sbin
## /sbin => system binary , admin/ root level commands are stored here like arp , ifconfig
## /opt => optional spplication software packages 
## /lib => shared libraries
## /dev => info of devices and drivers are stored here
## /mnt => The typical mount point for the user-mountable devices such as floppy drives and CDROM
## /proc => managed by linux to keep track of running processes. Virtual file system that provides system statistics
## /usr => Directory contains subdirectories with source code, programs, libraries, documentation, etc.
## /usr/sbin => system binaries for admin/ root level commands
## /usr/local/bin => user level commands
## /tmp => temporary files are stored here
## /boot => files used during the boot process along with the Linux kernel itself , dont modify

# Types of files in linux

### Files are mainly categorized into 3 parts:
## 1. Regular files
#### Regular or ordinary files store data of various content types such as text, audio, video, images, scripts, and programs. There are hundreds of content types. In Linux, regular files can be created with or without an extension.
## 2. Directory Files
#### Directory files are used to store other files and directories. They are also known as folders.
## 3. Special files
#### Linux treats all hardware devices (such as hard drives, printers, monitors, terminal emulators, and CD/DVD drives) as special files. This means that an application program can access and use files and devices in the same way. This feature makes developing programs in Linux easier and flexible.Linux places all special files or device files under the /dev directory.
### file Types in Linux  (using ls -l )
#### 1. Link file -> l
#### 2. Block file -> b
#### 3. Character file -> c
#### 4. Socket file -> s
#### 5. Directory file -> d
#### 6. Regular file -> -

#### root has every access to all files by default and any user cant change it

#### permission matrix
#### 7 = all rights
#### 6 = read and write
#### 5 = read and execute
#### 4 = read only
#### 3 = execute and write
#### 2 = write only
#### 1 = execute only
#### 0 = no rights

#### “r” is for read, “w” is for write, and “x” is for execute

#### 			  owner    group   size   date/time     file/dir name
#### drwxrwxr-x 4 avenger avenger  4096   Dec 27 12:07  t1

#### 	4 => permission matrix




#### Execute is the ability to run a file as an executable program.
#### Execute is the ability to cd into the directory and to view the metadata of the files inside the directory using ls command.

### 1]  Change file access mode (using chmod[change mode])

#### using numeric values
#### To give access as following use according to permission matrix

####		 rwx = 7 = all rights
####		 rw- = 6= read and write
####		 r-x = 5= read and execute
####		 r-- = 4 = read only
####		 -wx  =3 = execute and write
####		 -w-  = 2 = write only
####		 --x = 1 = execute only
####		 ---  = 0 = no rights
##### 3 consecutive numbers for users as =>  owner(u)    group-where-owner-belong(g)     other-users(o)

```bash
chmod 755 filename/dirname         # 755 means owner has all rights , group has read and execute and other has read and execute
# best practice , use descending only like 760 not 572 or 510 not 572
```

#### Using Alphabates and Signs with help of UGO 

####	the minus sign (-), which means “remove these permissions”
####	the plus sign (+), which means “add these permissions”
####	the equals sign (=), which means “change the permissions to exactly these”

```bash
chmod +x file/dir     # give execute persmission to all 3 types  , also chmod a+x file/dir can be used for  simplicity a for all
chmod -r file/dir     # remove read permission from all 3 types 
# UGO => u for user , g for group , o for other 
chmod ug+w file/dir   # give w access to owner(u) and group(g)
chmod u-wx,o-rwx file/dir
chmod u=rwx,g=rw,o=rw file/dir   # same as 766

```

### Ping

#### Ping is a computer network administration software utility used to test the reachability of a host on an Internet Protocol (IP) network. It measures the round-trip time for messages sent from the originating host to a destination computer that are echoed back to the source.

### Dig 

#### dig is a network administration command-line tool for querying Domain Name System (DNS) name servers. It performs DNS lookups and displays the answers that are returned from the name server that was queried.(Domain Information Groper)
#### 53 port is used by DNS so dig command uses ip:53 to query DNS server



```bash

dig www.google.com  # to get DNS info of google.com

dig -x www.google.com  # to get DNS info of google.com in reverse order

dig www.google.com +trace  # to get DNS info of google.com with trace

dig www.google.com +trace +noall +answer  # to get DNS info of google.com with trace and only answer

dig www.google.com +trace +noall +answer +stats  # to get DNS info of google.com with trace and only answer and stats

```

### Traceroute

#### traceroute is a computer network diagnostic tool for displaying the route (path) and measuring transit delays of packets across an Internet Protocol (IP) network. Traceroute determines the intermediate gateways (routers) a packet traverses to reach its destination. It is a command-line tool available on most Unix-like operating systems, including Linux, macOS, and BSD.

```bash

traceroute www.google.com  # to get route info of google.com

traceroute -I www.google.com  # to get route info of google.com with icmp

```

### Nslookup

#### nslookup is a computer network administration command-line tool available for many operating systems for querying the Domain Name System (DNS) to obtain domain name or IP address mapping or for any other specific DNS record.

```bash

nslookup www.google.com  # to get DNS info of google.com

nslookup -type=mx www.google.com  # to get DNS info of google.com with type mx (mail exchange)

```

### Netstat

#### netstat is a command-line network utility for displaying network connections for Transmission Control Protocol (TCP), User Datagram Protocol (UDP), and Internet Protocol (IP) network protocols. It is a standard feature of most Unix-like operating systems, including Linux, macOS, and BSD.

```bash

netstat     # to get all network connections

netstat -tulpn  # to get all tcp/udp listening ports

```


### Curl

#### curl is a command-line tool for transferring data using various network protocols. It is a computer software package providing a library and command-line tool for transferring data using various protocols. The command is designed to work without user interaction. As such, it is ideal for use in scripts, cron jobs, terminals without X-Windows support, etc.  (Client URL)


```bash

curl www.google.com  # to get html of google.com

curl -I www.google.com  # to get header of google.com

curl -I -X POST www.google.com  # to get header of google.com with post request

curl -I -X POST -d "name=abc" www.google.com  # to get header of google.com with post request and data

curl -X POST -d "name=abc" -H "Content-Type: application/json" www.google.com  # to get header of google.com with post request and data and header

```


### wget

#### wget is a free software package for retrieving files using HTTP, HTTPS, FTP and FTPS the most widely-used Internet protocols. It is a non-interactive commandline tool, so it may easily be called from scripts, cron jobs, terminals without X-Windows support, etc. (World Wide Web get)


```bash

wget www.google.com  # to get html of google.com

wget -O google.html www.google.com  # to get html of google.com and save it as google.html

wget -O google.html www.google.com -o wget.log  # to get html of google.com and save it as google.html and save wget log in wget.log

```

### ssh

#### ssh is a cryptographic network protocol for operating network services securely over an unsecured network. Typical applications include remote command-line, login, and remote command execution, but any network service can be secured with SSH. (Secure Shell)

```bash

ssh user@host  # to ssh to host with user at port 22

ssh user@host -p 2222  # to ssh to host with user and port 2222

ssh user@host -p 2222 -i ~/.ssh/id_rsa  # to ssh to host with user and port 2222 and private key

```

### scp

#### scp is a command-line utility for copying files between hosts on a network. It uses the SSH protocol for data transfer and provides the same authentication and same level of security as ssh. (Secure Copy)

```bash

scp file user@host:/home/user  # to copy file to host with user 

scp file user@host:/home/user -P 2222  # to copy file to host with user and port 2222

scp file user@host:/home/user -P 2222 -i ~/.ssh/id_rsa  # to copy file to host with user and port 2222 and private key

```

### rsync

#### rsync is a software application for Unix-like computer systems that provides fast incremental file transfer. It is a file transfer program capable of efficient remote update via a fast differencing algorithm. (Remote Sync)

```bash

rsync -avz file user@host:/home/user  # to copy file to host with user

rsync -avz file user@host:/home/user -e "ssh -p 2222"  # to copy file to host with user and port 2222

```


### iptables

#### iptables is a user-space utility program that allows a system administrator to configure the tables provided by the Linux kernel firewall (implemented as different Netfilter modules) and the chains and rules it stores. (IP Tables)

```bash

iptables -L  # to list all iptables

iptables -L -n  # to list all iptables with ip address

```

### tcpdump

#### tcpdump is a command-line network sniffer. It is used to display TCP/IP and other packets being transmitted or received over a network to which the computer is attached. It is a standard feature of most Unix-like operating systems, including Linux, macOS, and BSD. (TCP Dump)

```bash

tcpdump  # to dump all packets

```


### nmap

#### nmap is a free and open-source network scanner created by Gordon Lyon. Nmap is used to discover hosts and services on a computer network by sending packets and analyzing the responses. (Network Mapper)

```bash

nmap www.google.com  # to scan google.com

```
